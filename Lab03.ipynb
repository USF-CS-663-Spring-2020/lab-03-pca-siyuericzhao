# 1.WDBC Data Set

import pandas as pd
df = pd.read_csv('downloads/wdbc.data', header=None)
df

```python
column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
'mean smoothness', 'mean compactness', 'mean concavity',
'mean concave points', 'mean symmetry', 'mean fractal dimension',
'radius error', 'texture error', 'perimeter error', 'area error',
'smoothness error', 'compactness error', 'concavity error',
'concave points error', 'symmetry error',
'fractal dimension error', 'worst radius', 'worst texture',
'worst perimeter', 'worst area', 'worst smoothness',
'worst compactness', 'worst concavity', 'worst concave points',
'worst symmetry', 'worst fractal dimension']

df.columns = column_labels
df
```

```python
df['diagnosis'].replace('B','Benign', inplace=True)
df['diagnosis'].replace('M','Malignant', inplace=True)
df
```

```python
feature_df = df.iloc[:,2:]
label_df = df.iloc[:,1]
feature_df
```

```python
from sklearn.preprocessing import StandardScaler

#x = feature_ef.values
#x = StandardScaler().fit_transform(x)

x = StandardScaler().fit_transform(feature_df)
x
```

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
principal_breast_df = pd.DataFrame(data = principalComponents_breast,columns = ['PC1', 'PC2'])
```

```python
pca_breast.explained_variance_ratio_
```

we can see the two principle components explains almost sixty percent of the total variance.

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()

plt.xlabel('PC1')
plt.ylabel('PC2')
targets = ['Benign','Malignant']
colors = ['b','r']
for target, color in zip(targets, colors):
    indicesToKeep = label_df == target
    plt.scatter(principal_breast_df.loc[indicesToKeep, 'PC1'],
                principal_breast_df.loc[indicesToKeep, 'PC2'],
                c = color)
```

```python
pca_gi_train_all = PCA().fit(x)
evr = pca_gi_train_all.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
```

# 2. GISETTE

```python
#Load Gisette trained Data
gisette_train_df = pd.read_csv('Downloads/gisette_train.data', header=None, delim_whitespace=True)
gisette_train_df
```

```python
gisette_train_label_df = pd.read_csv('Downloads/gisette_train.labels', header=None, delim_whitespace=True)
```

```python
gisette_train_label_df
```

```python
gisette_train_label_df = gisette_train_label_df.iloc[:, 0]
gisette_train_label_df
```

```python
#Normalizing the train set
gisette_train_x = StandardScaler().fit_transform(gisette_train_df)
gisette_train_x
```

```python

np.mean(gisette_train_x), np.std(gisette_train_x)
```

As we can see from the above, the mean is very close to zero, so it is basically zero, and also the standard deviation very close to 1, therefore , we can see a normal distribution.

```python
pca_gisette = PCA(n_components=4)
```

```python
pca_gisette_data = pca_gisette.fit_transform(gisette_train_x)
```

```python
pca_gisette_data
```

```python
#put the PCA in to a data frame for us to better visualize
pca_gisette_df = pd.DataFrame(pca_gisette_data, columns=['PC1', 'PC2', 'PC3', 'PC4'])
pca_gisette_df
```

```python
pca_gisette.explained_variance_ratio_
```

We can see that from PC1 through PC4, the pricinciple component explains variances about 3%,2%,1.5% and 1.5%, compare to the WEBC data set, which PC1 and PC2 together explained almost 60% of the total variance. It is reasonable because there exist much dimension of date in this Gisette data set and thus multi dimensions between each data points.

```python
#List the first 50 variance and see the how each of them when combined explains the total variance
pca_gi_train_all = PCA().fit(gisette_train_x)
evr_gisette = pca_gi_train_all.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr_gisette[:50])))
```

```python
plt.plot(np.cumsum(evr_gisette))
plt.xlabel('Number of components')
```

From the above graph we can see that, from the number 2000 to 3000 demensions of to 5000 dimensions, if we take the number of components in between 2000 to 3000 we get above 90% and close to 100% of the data information, thus it is possible for us to reduce the data dimension to 2000.

```python
#Graph the first two principle component of train set
%matplotlib inline
plt.figure()
plt.xlabel('pc1')
plt.ylabel('pc2')
targets = [1, -1]
colors = ['b', 'r']
for target, color in zip(targets, colors):
    indicesToKeep = gisette_train_label_df == target
    plt.scatter(pca_gisette_df.loc[indicesToKeep, 'PC1'],
                pca_gisette_df.loc[indicesToKeep, 'PC2'],
                c=color)
```

we can see that there is a great deal of overlapping between the two components. Thus the isn't much to interprete about the first two components and the explanatory power of the first two component is very limited.
